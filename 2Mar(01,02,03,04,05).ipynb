```python
#1-Matplotlib is a popular data visualization library in Python. 
#It provides a wide range of tools and functions for creating various 
#types of plots, charts, and graphs. Matplotlib's pyplot module is a 
#collection of functions that provides a MATLAB-like interface for creating plots
#1-Line plot
import matplotlib.pyplot as plt
import numpy as np

x = np.linspace(0, 10, 100)
y = np.sin(x)

plt.plot(x, y)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Sine Wave')
plt.show()

```

```python
import matplotlib.pyplot as plt
import numpy as np

x = np.random.rand(100)
y = np.random.rand(100)
colors = np.random.rand(100)
sizes = 1000 * np.random.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.5)
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Scatter Plot')
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np

data = np.random.randn(1000)

plt.hist(data, bins=30)
plt.xlabel('Values')
plt.ylabel('Frequency')
plt.title('Histogram')
plt.show()
```

```python
import matplotlib.pyplot as plt

labels = ['A', 'B', 'C', 'D']
sizes = [30, 25, 15, 30]
colors = ['red', 'blue', 'green', 'yellow']

plt.pie(sizes, labels=labels, colors=colors, autopct='%1.1f%%')
plt.title('Pie Chart')
plt.show()
```

```python
#Q-3:The subplot() function in Matplotlib is used to create multiple 
#subplots within a single figure. It allows you to organize and display 
#multiple plots in a grid-like structure, making it easier to 
#compare and analyze different datasets or visualizations.
import numpy as np
import matplotlib.pyplot as plt

x = np.array([0, 1, 2, 3, 4, 5])

# Line 1
y1 = np.array([0, 100, 200, 300, 400, 500])
plt.subplot(2, 2, 1)  # Create a 2x2 grid of subplots, select the first subplot
plt.plot(x, y1)
plt.title('Line 1')

# Line 2
y2 = np.array([50, 20, 40, 20, 60, 70])
plt.subplot(2, 2, 2)  # Select the second subplot
plt.plot(x, y2)
plt.title('Line 2')

# Line 3
y3 = np.array([10, 20, 30, 40, 50, 60])
plt.subplot(2, 2, 3)  # Select the third subplot
plt.plot(x, y3)
plt.title('Line 3')

# Line 4
y4 = np.array([200, 350, 250, 550, 450, 150])
plt.subplot(2, 2, 4)  # Select the fourth subplot
plt.plot(x, y4)
plt.title('Line 4')

plt.tight_layout()  # Improve the spacing between subplots
plt.show()
```

```python
#Q-4:A bar plot, also known as a bar chart,
#is a type of chart that uses rectangular bars to 
#represent data. It is commonly used to compare and 
#display categorical data or to show the relationship 
#between a category and a numeric value. 
#The length of each bar corresponds to the magnitude of the data it represents.

#Bar plots are used for various purposes, including:

#Comparing data across different categories: Bar plots make it easy to compare the 
#values of different categories and identify trends or variations.

#Visualizing frequency or count data: Bar plots can display the frequency 
#or count of categorical variables, allowing for easy interpretation and analysis.

#Showing rankings or ordering: Bar plots can effectively represent rankings 
#or ordering of categories based on their values.
import numpy as np
import matplotlib.pyplot as plt

company = np.array(["Apple", "Microsoft", "Google", "AMD"])
profit = np.array([3000, 8000, 1000, 10000])

# Bar Plot
plt.bar(company, profit)
plt.xlabel('Company')
plt.ylabel('Profit (in millions)')
plt.title('Profit of Companies')
plt.show()

# Horizontal Bar Plot
plt.barh(company, profit)
plt.xlabel('Profit (in millions)')
plt.ylabel('Company')
plt.title('Profit of Companies')
plt.show()
```

```python
#Q-5:A box plot, also known as a box-and-whisker plot, 
is a statistical visualization tool used to display the distribution 
of a dataset through its quartiles. It provides a summary of the minimum, first quartile (25th percentile), median (50th percentile), third quartile (75th percentile), and maximum values.

A box plot consists of the following elements:

A rectangular box, which represents the interquartile range (IQR),
encompassing the middle 50% of the data.
A line inside the box, indicating the median value.
Two vertical lines extending from the box, known as whiskers, which represent the minimum and maximum values within a defined range.
Outliers, depicted as individual points lying outside the whiskers, indicating data points that significantly deviate from the rest.
```
